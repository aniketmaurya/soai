# Now, let's import the module which we're gonna use.

```python
from fastai.vision import *
```

`fastai` has a method to download datasets in a line.

```python
path = untar_data(URLs.DOGS)
```

### After downloading the dataset

1.   Define transformations https://docs.fast.ai/vision.transform.html
2.   Load the data
3.   Normalize the data



Image augmentation can be used in case we have less amount of data or we are overfitting.
But it can't be used everytime. 6

![Cat img augmentation](https://cdn-images-1.medium.com/max/780/1*5RXJ2OlVJp4_sd1XzhK6pQ.png =550x)


![Normalize](https://github.com/aniketmaurya/soai/blob/master/pics/Screenshot%20from%202019-03-30%2022-10-02.png?raw=true =750x)

Suppose for _House Price Prediction_ we have **number of bedrooms** and **area of house** then _number of bedrooms_ are generally less than 5 and _area_ can be 100 square metres.

So, the our two attributes have totally different scale. We normalize data to bring it to a similar range.

After normlization, generally all the values lies in range of 0 to 1.



```python
tfms = get_transforms()
data = ImageDataBunch.from_folder(path, 'train', size=128, ds_tfms=tfms)
data.normalize(imagenet_stats)
```

### Before proceeding further it is always a good practice to visualize your data

```python
data.show_batch(row=3, figsize=(10,10))
```

## Now we're ready to create our DL model

Since , we creating a CNN we use `fastai`'s method `cnn_learner`

```python
learn = cnn_learner(data, models.resnet34, metrics = accuracy)
```

## Let's find our learning rate

#### We all have some learning rate. Right?

#### If we study too fast then we understand the topic initially but after some time don't get anything ğŸ˜•.

#### If we study too slow then we learn everything but we have to learn before exams ğŸ˜›

### There is a term _learning Rate_ in ML/DL which is exactly same.

Let's find the best learning rate for our model.

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=1e-033)
```

## We can plot confusion matrix and see what are those images on which our model didn't work well

https://docs.fast.ai/vision.learner.html#ClassificationInterpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(6,6))
```

```python
interp.plot_confusion_matrix()
```

```python
img = learn.data.train_ds[0][0]
learn.predict(img)[0]
```
